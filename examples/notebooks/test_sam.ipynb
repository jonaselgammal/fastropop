# Skymap Smoke Test

Minimal package-based skymap generation example.


```python
import fastropop
from fastropop import SemiAnalyticPopulation
```

```python
pop = SemiAnalyticPopulation()
Nbinaries_mean = pop.compute_Nbinaries(verbose=True)
Nbinaries_draw = int(pop.generate_poisson_realization(1, key=0, verbose=True)[0])
```

```python
Nside = 16
skymaps_tot, skymaps_plus, skymaps_cross = pop.generate_skymaps(
    Nbinaries=min(Nbinaries_draw, 100000),
    Nside=Nside,
    batch_size=100000,
    key=1,
    verbose=True,
)
skymaps_tot.shape, skymaps_plus.shape, skymaps_cross.shape
```
